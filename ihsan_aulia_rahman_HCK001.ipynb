BMR Male

#**IV. Solving Problems**

##**A. Creating Functions**

> Create two explicit functions both takes male and female BMR calculation for then to be used in computation.

```python
def BMR_M(BW, BH, A):
  BMR = 66 + (13.7 * BW) + (5 * BH) - (6.8 * A)
  return BMR

def BMR_F(BW, BH, A):
  BMR = 655 + (9.6 * BW) + (1.8 * BH) - (4.7 * A)
  return BMR
```

##**B. Creating New Column**

> Create new column named BMR comprised of results of the base metabolic rate (BMR) calculation respective  each customer's gender. 

```python
BMR_List = []
for i in range(0,len(df)):
  if df['Gender'][i] == 'Male':
    BMR_List.append(BMR_M(df['Weight (kg)'][i], df['Height (cm)'][i], df['Age'][i]))
  else:
    BMR_List.append(BMR_F(df['Weight (kg)'][i], df['Height (cm)'][i], df['Age'][i]))

df['Intake'] = BMR_List
df
```

##**C. Drawing Average**

>  Calculate the average body weight for male gender with age more than equal to 24 years to less than equal to 35.

```python
Average_BW = df.loc[(df['Gender'] == 'Male') & (df['Age']>=24) & (df['Age']<=35)].agg({'Weight (kg)':'mean'})
```
